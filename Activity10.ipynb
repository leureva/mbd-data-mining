```python
%pip install mlxtend --upgrade
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
```

```python
path = "https://raw.githubusercontent.com/leureva/mbd-data-mining/main/Activity10_mycart.csv"
df = pd.read_csv(path,
                 header=None
                )
df.head()
```

```python
ndf=df.applymap(str)
ndf 
```

```python
trans=ndf.values.tolist()
trans
```

```python
te=TransactionEncoder()
te_ary=te.fit(trans).transform(trans)
te_ary
```

```python
teDF=pd.DataFrame(te_ary,
                 columns=te.columns_
                 )
teDF.drop('nan', axis=1,inplace=True)
teDF
```

```python
teDF.columns.tolist()
```

```python
top10=teDF.head(10)
ndf=top10.copy()
ndf
```

```python
teDF.shape
```

```python
ft=fpgrowth(teDF,
                        min_support=0.008,
                        use_colnames=True)
ft
```

```python
as_ruls=association_rules(ft,
                         metric='lift',
                         min_threshold=3)
as_ruls
```
